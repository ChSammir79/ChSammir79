```r
# Cargar los datos desde el archivo CSV actualizado
df <- read.csv("Datos.csv")

# Instalar dplyr si no lo tienes (comentar si ya está instalado)
# install.packages("dplyr")

# Cargar la librería dplyr
library(dplyr)

# Separar los datos por grupo
control <- filter(df, Grupo == "Control")$Glucosa
experimental <- filter(df, Grupo == "Experimental")$Glucosa
experimental_nuevo <- filter(df, Grupo == "Experimental_Nuevo")$Glucosa

# Test de normalidad Shapiro-Wilk para el grupo Control
shapiro_test_control <- shapiro.test(control)

# Test de normalidad Shapiro-Wilk para el grupo Experimental
shapiro_test_experimental <- shapiro.test(experimental)

# Test de normalidad Shapiro-Wilk para el grupo Experimental Nuevo
shapiro_test_experimental_nuevo <- shapiro.test(experimental_nuevo)

# Mostrar resultados
shapiro_test_control
shapiro_test_experimental
shapiro_test_experimental_nuevo
```

```r
library(ggplot2)

# Crear histograma para el grupo Control
ggplot(df %>% filter(Grupo == "Control"), aes(x = Glucosa)) +
  geom_histogram(binwidth = 5, color = "black", fill = "blue", alpha = 0.7) +
  labs(title = "Histograma de Glucosa - Grupo Control", x = "Glucosa", y = "Frecuencia") +
  theme_minimal()

# Crear histograma para el grupo Experimental
ggplot(df %>% filter(Grupo == "Experimental"), aes(x = Glucosa)) +
  geom_histogram(binwidth = 5, color = "black", fill = "green", alpha = 0.7) +
  labs(title = "Histograma de Glucosa - Grupo Experimental", x = "Glucosa", y = "Frecuencia") +
  theme_minimal()

# Crear histograma para el grupo Experimental Nuevo
ggplot(df %>% filter(Grupo == "Experimental_Nuevo"), aes(x = Glucosa)) +
  geom_histogram(binwidth = 5, color = "black", fill = "red", alpha = 0.7) +
  labs(title = "Histograma de Glucosa - Grupo Experimental Nuevo", x = "Glucosa", y = "Frecuencia") +
  theme_minimal()
```
